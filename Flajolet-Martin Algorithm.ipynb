```python
import hashlib
from typing import List
from statistics import median

class FlajoletMartin:
    def __init__(self, num_hash_functions):
        # Số lượng hàm băm được sử dụng trong thuật toán
        self.hash_functions = num_hash_functions
        # Danh sách chứa các hàm băm
        self.hashes = self.generate_hashes()

    def generate_hashes(self):
        # Tạo danh sách các hàm băm
        return [self.HashFunction(i) for i in range(self.hash_functions)]

    def uniques(self, string):
        # Danh sách để lưu trữ giá trị r trung bình từ mỗi hàm băm
        average_r = []

        # Tách chuỗi thành danh sách các từ
        stream = string.split()

        # Lặp qua từng hàm băm
        for i in range(self.hash_functions):
            # Tìm giá trị r lớn nhất cho mỗi từ trong chuỗi
            sum_r = max(self.get_first_zero(self.hashes[i].get_hash(word)) for word in stream)
            average_r.append(sum_r)

        # Sắp xếp giá trị r và tính giá trị trung bình
        average_r.sort()
        r = 0

        avg_mid = len(average_r) // 2

        if len(average_r) % 2 == 0:
            r = (average_r[avg_mid] + average_r[avg_mid + 1]) / 2
        else:
            r = average_r[avg_mid + 1]

        # Ước lượng số lượng phần tử duy nhất và trả về
        return int(2**r)

    def get_first_zero(self, val):
        # Trả về vị trí của bit đầu tiên không đặt trong giá trị được đưa vào
        if val == 0:
            return 0
        pos = 0

        for i in range(64):
            if (val & (1 << i)) == 0:
                pos += 1
            else:
                break
        return pos

    class HashFunction:
        def __init__(self, loop):
            # Số lần lặp trong hàm băm
            self.num_loop = loop

        def get_hash(self, s):
            # Hàm băm
            hash_val = 0
            i = 0

            # Lặp để áp dụng hàm băm theo số lần lặp đã cho
            while i < self.num_loop:
                if isinstance(s, str):
                    hash_val = hash(s) * (i + 1)
                    s = str(hash_val)
                elif isinstance(s, (int, float)):
                    hash_val = hash(str(s)) * (i + 1)
                    s = str(hash_val)
                else:
                    t = str(s)
                    hash_val = hash(t) * (i + 1)
                    s = str(hash_val)

                i += 1

            return 13 + 7 * abs(hash_val)
```

```python
class FMMapper:
    def __init__(self, num_hash_functions):
        # Tạo một đối tượng FlajoletMartin với số lượng hàm băm đã cho
        self.fm = FlajoletMartin(num_hash_functions)

    def map(self, input_string):
        # Áp dụng thuật toán Flajolet-Martin cho chuỗi đầu vào
        stream = input_string.split()
        distinct = self.fm.uniques(input_string)
        return distinct
if __name__ == "__main__":
    # Example usage
    # Đọc dữ liệu từ tệp txt
    with open('N:\wordcount.txt', 'r') as file:
        input_data = file.read()

    # Chia dữ liệu thành từng dòng
    lines = input_data.strip().split('\n')

    for line in lines:
        # Tách ID và stream
        id, stream = line.split(' ', 1)

        # Tạo một đối tượng FMMapper và áp dụng thuật toán
        fm_mapper = FMMapper(num_hash_functions=4)
        result = fm_mapper.map(stream)

        # In kết quả
        print(f"Đầu dòng ID: {id}, Phần tử ước tính: {result}")
```
